```python
class CustomDatasetFromImages(Dataset):
    def __init__(self, data_info):
        #    
        self.data_info = data_info
        
        #    rgb  
        self.image_arr = self.data_info.iloc[:,0]
        #    
        self.label_arr = self.data_info.iloc[:,1]
        
        #   -
        self.data_len = len(self.data_info.index)
        
        
#         self.class_values = np.unique(plt.imread(self.label_arr[index])[:,:,0]*255)
        
    def __getitem__(self, index):
        #          numpy- 
        #  600800 float-
        img = Image.open(self.image_arr[index])
        #    256256
        img = img.resize((256,256))
        img = np.asarray(img).astype('float')
        #     [0,1]
        img = torch.as_tensor(img)/255    
        #       -   pytorch
        img = img.permute(2,0,1).float()
        
        class_values = np.unique(plt.imread(self.label_arr[index])[:,:,0]*255)
        
        #   
        mask = np.asarray(plt.imread(self.label_arr[index]))[:,:,0]*255

        #            
        masks = [(mask == v) for v in class_values]
        #        
        mask = np.stack(masks, axis=-1).astype('float')
        #     
        mask = torch.as_tensor(mask)
        #     
        mask = mask.permute(2,0,1)
        
        #     256256
        #     interpolate
        ###        ,   interpolate
        ###  unsqueeze  ,  squeeze .
        # unsqueeze -   img c (256, 256, 3) -> (1, 256, 256, 3),
        mask = mask.unsqueeze(0)
        mask = F.interpolate(input=mask, size=256, mode='nearest')
        mask=mask.squeeze(0).squeeze(0)
        
        
        return (img, mask)

    def __len__(self):
        return self.data_len
```
